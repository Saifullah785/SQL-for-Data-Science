# **üß† What Are Database Keys?**


---

A DBMS key is an attribute  or set of attributes or set of column which helps us uniquely identify a row in a relation/table

**Keys are unique identifiers used to:**

Identify records (rows) in a table

Create relationships between different tables

Enforce data integrity (no duplicates, valid links)


üßæ Analogy: Think of a key like an ID card. It uniquely identifies a person (row) in a database (population).

# **why do we need keys**

1.To uniquely identify a row

2.To enforce Data integrity and constraints

3.To establish relationship between table

# **Types of Keys**

1.Super key

2.Candidate Key

3.primary Key

4.Alternate key

5.Foreign Key

6.Composite Key

7.Compound Key

8.Surrogate Key


# **ü¶∏ 1. Super Key ‚Äî All Possible Unique Identifiers**


---


**üìå What it is:**

A set of one or more columns that uniquely identify a record

Includes Primary Key, Composite Key, and any Candidate Key

üí° Super key = Candidate key + Extra columns (optional)

**Example:**

user_id

user_id + email

user_id + username + email ‚Üí All are super keys (but only minimal ones are candidate keys)

# **üß† 2. Candidate Key ‚Äî Potential Primary Keys**


---


**üìå What it is:**

the candidate key is a minimum number of column required to become a key

Any column(s) that can uniquely identify a row

A table can have multiple candidate keys

One of them becomes the Primary Key

**‚úÖ Real-life example:**

In a users table, both email and username may be unique

üßæ SQL Setup:
sql

CREATE TABLE users (

    user_id INT,

    email TEXT UNIQUE,

    username TEXT UNIQUE,

    PRIMARY KEY (user_id)

);

# **üîë 3. Primary Key ‚Äî Main Unique Identifier**


---



**üìå What it is:**

A column (or combination) that uniquely identifies each row.

Cannot be NULL

Must be unique

**‚úÖ Real-life example:**

Student ID in a student table

Order ID in an e-commerce system

**üßæ SQL Example:**

sql

CREATE TABLE students (

    student_id INT PRIMARY KEY,

    name TEXT,

    age INT

);

# **üîÅ 4. Alternate Key ‚Äî Leftover Candidate Key**


---


**üìå What it is:**

A candidate key that was not chosen as the primary key

**‚úÖ Real-life example:**

In the above users table:

email and username are candidate keys

If user_id is chosen as the primary key, then email and username become alternate keys

Candidate key - Primary key = Alternate key

# **üîó 5. Foreign Key ‚Äî Link Between Tables**


---



**üìå What it is:**

A column in Table B that refers to the Primary Key in Table A

Creates a relationship between two tables

Enforces referential integrity

**‚úÖ Real-life example:**

A student‚Äôs class_id in the students table referring to the class_id in the classes table

**üßæ SQL Example:**
sql


-- Classes Table (Parent)

CREATE TABLE classes (

    class_id INT PRIMARY KEY,

    class_name TEXT
);

-- Students Table (Child)

CREATE TABLE students (
    
    student_id INT PRIMARY KEY,
    
    name TEXT,
    
    class_id INT,
    
    FOREIGN KEY (class_id) REFERENCES classes
    
    (class_id)
);

# **üß© 6. Composite Key ‚Äî Combined Uniqueness**


---



**üìå What it is:**

A combination of two or more columns that uniquely identify a record.

Used when no single column is unique enough

**‚úÖ Real-life example:**

In a course_enrollment table:

student_id + course_id = uniquely identifies a course enrollment record

**üßæ SQL Example:**

sql

CREATE TABLE course_enrollments (

    student_id INT,

    course_id INT,

    enrollment_date DATE,

    PRIMARY KEY (student_id, course_id)

);

# **7.Compound Key ‚Äî Subtype of Composite Key**

**üìå What it is:**

A Composite Key where each column is also a Foreign Key

It creates uniqueness and connects two related tables

**‚úÖ Real-life Example:**

A student_courses table where:

Each row represents a student's enrollment in a course

student_id ‚Üí refers to students

course_id ‚Üí refers to courses

Together, they uniquely identify an enrollment

**üßæ SQL Example:**

sql

CREATE TABLE student_courses (

    student_id INT,

    course_id INT,

    enrollment_date DATE,

    PRIMARY KEY (student_id, course_id),

    FOREIGN KEY (student_id) REFERENCES
    
    students(student_id),
    
    FOREIGN KEY (course_id) REFERENCES courses(course_id)

);

üí° Compound Key = Composite Key + Foreign Keys

# **8.Surrogate Key ‚Äî Artificial Unique ID**

**üìå What it is:**

A column (usually auto-incremented) that has no business meaning

Used as a Primary Key instead of natural keys

Common in data warehouses and dimensional models

**‚úÖ Real-life Example:**

Instead of using email, username, or SSN as primary keys (all meaningful), we just assign:



user_id: 1, 2, 3, ...

**üßæ SQL Example:**

sql

CREATE TABLE employees (
   
    employee_id SERIAL PRIMARY KEY,  --
    
    Surrogate Key
   
    first_name TEXT,
   
    last_name TEXT,
   
    email TEXT

);

employee_id is a surrogate key

Has no meaning outside this table

Makes JOINs faster and safer

# **Cardinality of Relationships**


---

Cardinality of relationship is the number of instances in one entity which is associated to the number of instances in another.

For the relationship between Employee and Computer, it helps us answer question like how many computers can be allocated to an employee, can computers be shared between employees, can employees exist without being allocated a computer etc.e.g if 0 or 1 computer can be allocated to 0 or 1 employee then the cardinality of relationship between these two entities will be 1:1
