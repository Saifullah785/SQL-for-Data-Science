# **ðŸ”§ 1. DDL â€“ Data Definition Language**


---


"Build or change the structure of your tables"

ðŸ’¡ Common DDL Commands:

CREATE â€“ Make tables or databases

ALTER â€“ Change structure (add column, modify type)

DROP â€“ Delete a table/database

TRUNCATE â€“ Delete all records (faster than DELETE)

**ðŸ§ª SQL Examples:**

CREATE TABLE customers (
   
    id INT PRIMARY KEY,
   
    name TEXT,
   
    email TEXT UNIQUE

);

ALTER TABLE customers ADD COLUMN phone TEXT;

DROP TABLE old_customers;

# **ðŸ“Š 2. DML â€“ Data Manipulation Language**

"Work with the actual data"

ðŸ’¡ Common DML Commands:

INSERT â€“ Add new records

UPDATE â€“ Modify records

DELETE â€“ Remove records

**ðŸ§ª SQL Examples:**

INSERT INTO customers (id, name, email) VALUES (1, 'John', 'john@gmail.com');

UPDATE customers SET email = 'john.doe@gmail.com' WHERE id = 1;

DELETE FROM customers WHERE id = 1;

# **ðŸ“¤ 3. DCL â€“ Data Control Language**

"Control who can access the data"

ðŸ’¡ Common DCL Commands:

GRANT â€“ Give access

REVOKE â€“ Remove access

**ðŸ§ª SQL Example:**

GRANT SELECT, INSERT ON customers TO analyst_user;

REVOKE INSERT ON customers FROM analyst_user;

# **ðŸ” 4. TCL â€“ Transaction Control Language**

"Make operations permanent or cancel them"

ðŸ’¡ Common TCL Commands:

COMMIT â€“ Save changes

ROLLBACK â€“ Undo changes

SAVEPOINT â€“ Mark checkpoints

**ðŸ§ª SQL Example:**


BEGIN;

UPDATE accounts SET balance = balance - 100 WHERE id = 1;

UPDATE accounts SET balance = balance + 100 WHERE id = 2;

-- Error occurs here

ROLLBACK; -- All changes undone

-- or

COMMIT; -- Save permanently

# **ðŸ” 5. DQL â€“ Data Query Language**

"Query or fetch data"

ðŸ’¡ Common DQL Command:

SELECT â€“ Retrieve data

**ðŸ§ª SQL Example:**


SELECT name, email FROM customers WHERE id < 10;


| Command Type | Keyword Examples              | Purpose                   |
| ------------ | ----------------------------- | ------------------------- |
| DDL          | CREATE, ALTER, DROP, TRUNCATE | Define database structure |
| DML          | INSERT, UPDATE, DELETE        | Work with table data      |
| DCL          | GRANT, REVOKE                 | Set permissions           |
| TCL          | COMMIT, ROLLBACK, SAVEPOINT   | Control transaction flow  |
| DQL          | SELECT                        | Query data                |


# **ðŸ—‚ï¸ Part 2: How Professionals Use ER Diagrams Before Creating a Database**


---


ðŸŽ¯ What is an ER Diagram?

An Entity-Relationship (ER) Diagram is a visual map of your database. It shows:

Entities (tables like users, orders)

Attributes (columns like name, price)

Relationships (how tables are connected)

# **ðŸ’¼ Why Professionals Use ERDs Before SQL?**

| Without ERD               | With ERD                  |
| ------------------------- | ------------------------- |
| Messy and inconsistent DB | Clean, well-structured DB |
| Hard to understand schema | Easy to explain to team   |
| Risk of data redundancy   | Normalized structure      |


# **ðŸ§  Key Components of an ERD**

| Concept      | Description                    | SQL Equivalent           |
| ------------ | ------------------------------ | ------------------------ |
| Entity       | Real-world object (Table)      | `products`, `customers`  |
| Attribute    | Properties of entity (Columns) | `name`, `price`, `email` |
| Primary Key  | Unique identifier              | `id INT PRIMARY KEY`     |
| Foreign Key  | Link to another entity         | `FOREIGN KEY (user_id)`  |
| Relationship | Connection between entities    | JOINs based on keys      |
