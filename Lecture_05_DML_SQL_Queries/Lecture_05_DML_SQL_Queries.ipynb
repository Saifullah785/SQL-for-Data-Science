# **8.RETRIEVE**


---



**--All cols**

SELECT * FROM train
===================================================================

**--Filter cols**

SELECT Name,Sex,Survived from train

===================================================================

**--Alias As**

SELECT Name AS PassengerName , Sex AS Gender, Survived from train

===================================================================

**--Expression**

SELECT Name, SibSp +Parch as family from train

SELECT Name,Age +105 AS Current_age FROM train

===================================================================

**--Constant**

SELECT Name,100000 AS compensation FROM train

===================================================================

**--Distinct Single Col**

SELECT DISTINCT Sex FROM train

SELECT DISTINCT Embarked FROM train

SELECT DISTINCT Pclass,Embarked FROM train

===================================================================

**--Comparision Operator**

SELECT * FROM train WHERE Survived = 0

SELECT * FROM train WHERE Age >50

===================================================================

**--AND/OR/BETWEEN**

SELECT * FROM train WHERE Pclass = 3 AND Survived =0

SELECT * FROM train WHERE Age BETWEEN 10 AND 15

===================================================================

**--IN/NOT IN**

SELECT name,genre FROM movies WHERE genre IN ('Action','Horror','Drama')

SELECT name,genre FROM movies WHERE genre NOT IN ('Action','Horror','Drama')

===================================================================

**--Wildcards**

SELECT name FROM movies WHERE name LIKE 'A%'

SELECT name FROM movies WHERE name LIKE '%man%'

SELECT name FROM movies WHERE name LIKE '%a'

SELECT name,star FROM movies WHERE star LIKE '%khan%' OR star LIKE '% kapoor%'

SELECT name FROM movies WHERE name LIKE 'a___'

===================================================================

**--Order of Query execution**

FROM JOIN WHERE GROUP BY HAVING SELECT DISTINCT ORDER BY

# **11.function**


---



**--ABS**

SELECT name,ABS((gross - budget)) AS profit FROM movies

===================================================================

**--ROUND**

SELECT name, ROUND(runtime/60) AS runtime_hrs FROM movies

SELECT name, ROUND((runtime/60),2) AS runtime_hrs FROM movies

===================================================================

**--CEIL**

SELECT name, CEIL((runtime/60)) AS runtime_hrs FROM movies

===================================================================

**--FLOOR**

SELECT name, FLOOR((runtime/60)) AS runtime_hrs FROM movies

===================================================================

**--UPER/LOWER**

SELECT UPPER (name) FROM movies

===================================================================

**--CONCAT**

SELECT name,CONCAT(star,'',director)AS crew FROM movies

===================================================================

**--LENGTH**

SELECT name,LENGTH(name)as length from movies

===================================================================

**--SUBSTR(string,start(1),Length)**

SELECT name ,SUBSTR(name,1,5)AS short FROM movies

===================================================================

# **13.GROUPING DATA**


---



**--GROUP BY**

SELECT star,COUNT(*) AS num_movies FROM movies GROUP BY star ORDER BY

num_movies DESC LIMIT 5

===================================================================

SELECT genre,SUM(gross - budget) AS total_profit from movies GROUP BY genre

ORDER BY total_profit DESC LIMIT 5

===================================================================

SELECT director,AVG(gross - budget) AS avg_profit FROM movies GROUP BY  

director ORDER BY avg_profit DESC LIMIT 5

===================================================================

SELECT name,budget FROM movies GROUP BY name ORDER BY budget DESC LIMIT 5

===================================================================

SELECT star,director, SUM(gross - budget) AS profit FROM movies GROUP BY star,director ORDER BY profit DESC	LIMIT 5

===================================================================

SELECT star,genre, SUM(gross - budget) AS profit FROM movies GROUP BY star,genre ORDER BY profit DESC LIMIT 5

===================================================================

SELECT star,SUM(gross - budget) AS total_profit FROM movies GROUP BY star ORDER BY total_profit DESC LIMIT 10

# **15. Having**


---



SELECT star,AVG(score)AS opening

FROM movies

GROUP BY star

HAVING opening>7

ORDER BY opening DESC
