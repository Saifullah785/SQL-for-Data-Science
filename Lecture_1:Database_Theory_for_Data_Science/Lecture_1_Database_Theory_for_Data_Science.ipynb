# **ðŸ§  Concept: What Is a Database?**


---



A database is a structured way of storing, retrieving, and managing data.


It stores data in tables (like Excel sheets).


Each table contains rows (records) and columns (fields).


Tables can be linked using primary keys and foreign keys.
